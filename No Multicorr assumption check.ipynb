```python
def correlation_strong_corr(data, thresh):
    setm=set()
    correlation_matrix=data.corr()
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if abs(correlation_matrix.iloc[i,j])> thresh:
                colname=correlation_matrix.columns[i]
                setm.add(colname)
    return setm
```

```python
correlation_strong_corr(df_x, 0.7)
```

```python
def correlation_rand_dist(data, thresh):
    setm=set()
    correlation_matrix=data.corr()
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if abs(correlation_matrix.iloc[i,j])< thresh:
                colname=correlation_matrix.columns[i]
                setm.add(colname)
    return setm
```

```python
correlation_rand_dist(df_x, 0.3)
```

```python
Total_col_out=len(correlation_strong_corr(df_x, 0.7))+len(correlation_rand_dist(df_x, 0.3))
Total_col_out
```
